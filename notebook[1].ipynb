# k-NN (Public)

```
import random
from random import sample

knn = tm.copy()

def distance(list1, list2):
    return np.sqrt(sum(np.subtract(list1, list2)**2))

rishi_27 = []
for i in knn["Rishi Vote Share"]:
    if i >= 0.27:
        rishi_27.append("Yes")
    else:
        rishi_27.append("No")

knn["Rishi Over 27%?"] = rishi_27
knn
```

```
knn = knn[["CarProp", "BusProp", "TrainProp", "Rishi Over 27%?"]]
knn_train = knn.sample(12)
knn_test = knn.sample(3)
```

```
def knn_classify(test_row, train_table, k):
    del test_row[3]
    modified_train = knn_train[["CarProp", "BusProp", "TrainProp"]]
    modified_train = modified_train.reset_index()
    distances = []
    for i in np.arange(0, 13, 1):
        train_row = modified_train.iloc[i].to_list()
        distances.append(distance(train_row, test_row))
    knn_copy = knn_train.copy()
    knn_copy["Distance"] = distances
    knn_copy = knn_copy.sort_values(by="Distance")
    knn_copy = knn_copy.rest_index()
    j = k+1
    knn_take = knn_copy.iloc[0:j]
    knn_take1 = knn_take["Rishi Over 27%?"].count("Yes")
    if knn_take1 >= 5:
        return("Yes")
    else:
        return("No")
```
